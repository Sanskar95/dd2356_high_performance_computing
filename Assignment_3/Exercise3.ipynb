```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
students2 = {'data_size':                   [
      
   
  
  16777216,
  33554432,
  67108864,
 134217728,
 268435456,
 536870912,
1073741824],
            'time': [



0.001696527,
0.003426504,
0.007456739,
0.015369492,
0.031350830,
0.063348162,
0.127482116]}
student_data2 = pd.DataFrame(data=students2)
```

```python
students = {'data_size':          [
     
   
   
  16777216,
  33554432,
  67108864,
 134217728,
 268435456,
 536870912,
1073741824],
            'time': [



0.002764444,
0.005641551,
0.011562622,
0.023420081,
0.047042418,
0.094443722,
0.189157577]}
student_data = pd.DataFrame(data=students)
```

```python

x = student_data.data_size
y = student_data.time

x2 = student_data2.data_size
y2 = student_data2.time
```

```python
plt.scatter(x,y)
plt.scatter(x2,y2)
```

```python
model = np.polyfit(x, y, 1)
model2 = np.polyfit(x2, y2 , 1)
```

```python
print(model)
print(model2)
```

```python
predict = np.poly1d(model)
predict2 = np.poly1d(model2)
```

```python
x_lin_reg = [(2 ** i) for i in range(3,31)]

y_lin_reg = predict(x_lin_reg)

x_lin_reg2 = [(2 ** i) for i in range(3,31)]

y_lin_reg2 = predict2(x_lin_reg2)

f = plt.figure()
f.set_figwidth(12)
f.set_figheight(12)
plt.scatter(x, y)
plt.scatter(x2, y2)
plt.xlabel("Data Size")
plt.ylabel("Time")

plt.plot(x_lin_reg, y_lin_reg, c = 'b')
plt.plot(x_lin_reg2, y_lin_reg2, c = 'r')
plt.legend(["Intra Node", "Inter Node"])
```
